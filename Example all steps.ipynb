```python
from imageprocessing.resize import resize_all
from imageprocessing.flatten import flatten_images, keep_top_k_classes
from dr.dr import apply_dr
from dr.plot import plot_projection

# import ray
import os
import numpy as np
```

```python
# ray.shutdown()
# ray.init(memory=10e+9, object_store_memory=5e+9)

name = 'paintings'
size = 100
c = 10
```

```python
# Download data
crawl_the_soup(file_of_links = 'data/paintings/painting_dataset_2018.csv', 
              save_folder='data/paintings/of/')
```

```python

folder = 'data/{}/original_files/'.format(name)
resize_all(folder, 'data/{}/thumbnails/size{}/'.format(name,size), dim=[size,size])
```

```python

flatten_images(input_folder = 'data/{}/thumbnails/size{}/'.format(name,size),
               output_folder = 'data/{}/flatfiles/'.format(name), 
               filename = '{}_flat{}'.format(name,size))

```

```python
keep_top_k_classes(filename='data/{}/flatfiles/{}_flat{}.csv'.format(name,name, size), c=c)
```

```python
files = [name]

for f in files:
    input_file = 'data/{}/flatfiles/{}_flat{}_{}classes.csv'.format(f,f,size,c)
    print(input_file)
    output_folder = 'data/{}/dr_data/size{}/'.format(f,size)
    print(output_folder)
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for dr_name in ['PCA','TSNE', 'ISM', 'UMAP']:
        print(dr_name)
        
        if dr_name in ['PCA']:
            dr_data = apply_dr.remote(input_file =input_file,
                                      output_folder=output_folder, 
                                      dataset_name = f, 
                                      dr_name=dr_name)
            
        elif dr_name == 'TSNE':
            for p in [5,10,30,100]:
                #print(p)
                dr_data = apply_dr.remote(input_file =input_file,
                                      output_folder=output_folder, 
                                      dataset_name = f, 
                                      dr_name=dr_name,
                                      perplexity=p)   
        
        else:
            for n in [2,5,7,15]:
                #print(n)
                if dr_name == 'UMAP':
                    for d in [0.1, 0.5, 0.8]:
                        dr_data = apply_dr.remote(input_file =input_file,
                                      output_folder=output_folder, 
                                      dataset_name = f, 
                                      dr_name=dr_name, n_neighbors=n, min_dist=d)
                elif dr_name in ['LLE','SE']:
                    if n == 7:
                        dr_data = apply_dr.remote(input_file =input_file,
                                      output_folder=output_folder, 
                                      dataset_name = f,  n_neighbors=n)
                    else: 
                        pass
                    
                else:
                    pass
```
